#  Reduction of co<sub>2</sub> emission and the shift towards low-carbon sources of energy

**Disclaimer**: The data used were obtained from external sources. The exact origins of the data are not currently documented. The analysis presented here is done with the primary intention of learning and should not be concluded to be true without validation. 

Since the Industrial Revolution, the energy mix of most countries across the world has become dominated by fossil fuels. By the end of the twentieth century, the world was completely dependent on and rapidly depleting the planet's fossil fuels— limited 
resources such as coal, natural gas, and oil. However it was not until recently did we realize it's major consequences upon global climate, as well as human health. To reduce CO<sub>2</sub> emissions and local air pollution, there is a need of change in our energy dependencies, that is, shifting from fossil fuels to low-carbon sources of energy – nuclear and renewable technologies.

This analysis aims to study the CO<sub>2</sub> emission trend over the years and increasing usage of other sources of energy especially in the last decade. 

```python
# Importing the necessery libraries 
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn 

# Get the co2 emission data 
CO2_emission = pd.read_csv("co-emissions-per-capita.csv")
CO2_emission.head()
```

Variables - 
1. *Entity* - country name
2. *Code* - code name for the corresponding country
3. *Year* - year
4. Annual CO₂ emissions (per capita) - contribution of a persion towards CO<sub>2</sub> emission in the particular year

```python
CO2_emission.info()
```

```python
# co2 emission - world
CO2_emission_world = CO2_emission.groupby('Year')['Annual CO₂ emissions (per capita)'].sum().reset_index(name='Annual CO2 emmission (world)').sort_values(by=['Year'])
```

```python
# Plotting the line graph 

plt.figure(figsize=(8,5)) 

sns.set_style("whitegrid")
sns.lineplot(x="Year",
             y="Annual CO2 emmission (world)",
             data=CO2_emission_world
            )

plt.xlabel('Year')
plt.ylabel('co2 emission (per capita)')
plt.title('Annual CO2 emmission (world)')
plt.show()
```

From the graph above we can infer that the carbon emission started raise up in the year 1950. Let's take a closer look at after 1950 emission records:

```python
# co2 emission after 1950
CO2_emission_world_1950 = CO2_emission_world[CO2_emission_world["Year"] > 1949].sort_values(by=['Year'])
```

```python
plt.figure(figsize=(9,5)) 

sns.set_style("whitegrid")
sns.lineplot(x="Year",
             y="Annual CO2 emmission (world)",
             data=CO2_emission_world_1950)

plt.xlabel('Year')
plt.ylabel('co2 emission (per capita)')
plt.title('Annual CO2 emmission(world) after 1950')
plt.show()
```

We can draw few conclusions from this graph. We can see that the emission was highest in the year of 1990, also we can notice the sudden escalation and immediate fall around that year. Also, interestingly the curve falls from year 2010 and indicates significant drop in CO<sub>2</sub> emission. The reasons for this behaviour might include various factors, such as actions taken regarding net zero emissions and recent covid-19 pandamic. 

```python
# co2 emission after 2019 (World)

CO2_emission_2019 = CO2_emission[CO2_emission["Year"] > 2018].sort_values(by=['Year']) # After 2019
CO2_emission_2019_world = CO2_emission_2019.groupby('Year')['Annual CO₂ emissions (per capita)'].sum().reset_index(name='Annual CO₂ emissions (world)') * 1
CO2_emission_2019_world 
```

From the above figures we can see the overall decrease in carbon emission. Let's compare the carbon emissions decades apart. 

```python
CO2_emission_2020 = CO2_emission[CO2_emission["Year"] == 2020]['Annual CO₂ emissions (per capita)'].sum()
CO2_emission_2020 
```

```python
CO2_emission_1990 = CO2_emission[CO2_emission["Year"] == 1990]['Annual CO₂ emissions (per capita)'].sum()
CO2_emission_1990 
```

```python
decrease_in_CO2_emission = (CO2_emission_1990 - CO2_emission_2020) / CO2_emission_1990 * 100
decrease_in_CO2_emission
```

This concludes that the emission of greenhouse gasses have reduced roughly 9.4% since 1990. It is important to note these figures do not count the co2 gas that is already in the atmosphere, but the co2 gas released into the air each year. 

# Renewable energy sources 

Renewable energy resources such as wind and sunlights are naturally replenished at a higher rate than they are consumed, providing a reliable source for energy for a long timescale. As the world started shifting from usage of fossil fuels to renewable energy, there was a notable amount of decrease in the emission of CO<sub>2</sub>. 

```python
# Get the data 

renewable_energy = pd.read_csv("renewable-share-energy.csv")
renewable_energy.head()
```

Variables - 

1. *Entity* - country name
2. *Code* - code name for the corresponding country
3. *Year* - year
4. *Renewables (% equivalent primary energy)* - proportion of renewable energy in total energy consumption. 

```python
renewable_energy.info()
```

Many countries have been using solar and wind power to satisfy their energy needs for decades. Let's see how much of renewable energy each country uses with respect to it's total energy consumption:

```python
# Grouping by country 
renewable_energy_by_country = renewable_energy.groupby('Entity')['Renewables (% equivalent primary energy)'].sum().reset_index(name = 'per_country').sort_values(by = 'per_country')
```

```python
# Plotting the graph 
plt.figure(figsize=(7,7)) 

sns.set_style("whitegrid")
sns.barplot(x = 'Entity',
            y = 'per_country',
            data = renewable_energy_by_country[renewable_energy_by_country["per_country"] > renewable_energy_by_country["per_country"].quantile(0.95)],
            palette = "YlGnBu",
            width=0.5)

plt.xlabel('Year')
plt.ylabel('Renewables (% equivalent primary energy)')
plt.title('cumulative proportion of renewable energy (top countries)')
plt.show()
```

These are few of those countries which are using renewable energy sources at highest potential.<br> <br>
Let's just consider the top three:

```python
countries = ["Norway", "Brazil", "New Zealand"]
renewable_energy_top_countries = renewable_energy[renewable_energy["Entity"].isin(countries)]
renewable_energy_top_countries
```

```python
plt.figure(figsize=(9,5)) 

sns.set_style("whitegrid")
sns.lineplot(x="Year",
             y="Renewables (% equivalent primary energy)",
             data=renewable_energy_top_countries, 
             hue = "Entity",
            palette = "YlGnBu")

plt.xlabel('Year')
plt.ylabel('Renewables (% equivalent primary energy)')
plt.title('Renewable Energy Usage (NOR, BRA, NZL)')
plt.show()
```

```python
# Renewable Energy Usage in India 
plt.figure(figsize=(9,5)) 

sns.set_style("whitegrid")
sns.lineplot(x="Year",
             y="Renewables (% equivalent primary energy)",
             data=renewable_energy[renewable_energy["Entity"] == "India"]
            )

plt.xlabel('Year')
plt.ylabel('Renewables (% equivalent primary energy)')
plt.title('Renewable Energy Usage (IND)')
plt.show()
```

```python
renewable_energy_global = renewable_energy.groupby('Year')['Renewables (% equivalent primary energy)'].sum().reset_index(name='Renewables_global').sort_values(by=['Year'])
```

```python
plt.figure(figsize=(7,5))

sns.set_style("whitegrid")
sns.lineplot(x="Year",
             y="Renewables_global",
             data=renewable_energy_global
            )

plt.xlabel('Year')
plt.ylabel('Renewables (% equivalent primary energy)')
plt.title('Renewable Energy Usage (World)')
plt.show()
```

# Solar Energy 

Many traditional power plants, such as coal and natural gas power plants, burn fossil fuels to produce electricity. However Solar energy, on the other hand, does not involve burning fossil fuels and is a clean, renewable source of electricity, which is why it helps reduce CO2 emissions.

```python
solar_energy = pd.read_csv("solar-energy-consumption.csv")
solar_energy.head()
```

Variables - 

1. *Entity* - country name
2. *Code* - code name for the corresponding country
3. *Year* - year
4. *Electricity from solar (TWh)* - electricity produced from solar power in terawatt-hour 

```python
solar_energy.info()
```

```python
# Solar power produced by different entities 
solar_energy_by_entity = solar_energy.groupby('Entity')['Electricity from solar (TWh)'].sum().reset_index(name = 'per_country').sort_values(by = 'per_country')
solar_energy_by_entity = solar_energy_by_entity[solar_energy_by_entity["Entity"] != "World"]
```

```python
plt.figure(figsize=(17,12)) 

sns.set_style("whitegrid")
sns.barplot(x = 'Entity',
            y = 'per_country',
            data = solar_energy_by_entity[solar_energy_by_entity["per_country"] > solar_energy_by_entity["per_country"].quantile(0.965)], 
            width=0.5,
            palette = "YlGnBu")

plt.xlabel('Year')
plt.ylabel('Solar Energy produced (TWh)')
plt.title('Solar power produced by entities')
plt.show()
```

```python
# Total solar power produced (world)
solar_energy_world = solar_energy[solar_energy["Entity"] == "World"]
```

```python
plt.figure(figsize=(7,5))

sns.set_style("whitegrid")
sns.lineplot(x="Year",
             y="Electricity from solar (TWh)",
             data=solar_energy_world
            )

plt.xlabel('Year')
plt.ylabel(' Solar Energy produced (TWh)')
plt.title('Solar power produced (World)')
plt.show()
```

# Electronic vehicles 

Electric vehicles (EVs) are a promising solution to reduce CO2 emissions in the transportation sector. They run on electricity, which can be generated from renewable sources like solar and wind, significantly lowering the carbon footprint associated with conventional gasoline-powered vehicles.

```python
# Get the data 
EV_sales = pd.read_csv("IEA-EV-dataEV salesHistoricalCars.csv")
EV_sales.head()
```

```python
EV_sales.info()
```

```python
EV_sales = EV_sales[EV_sales["parameter"] == "EV sales"]
```

```python
EV_sales_world = EV_sales[EV_sales["region"] == "World"]
EV_sales_world.describe()
```

```python
# EV sales in the world - line plot 
plt.figure(figsize=(7,5))

sns.set_style("whitegrid")
sns.lineplot(x="year",
             y="value",
             data=EV_sales_world
            )

plt.xlabel('Year')
plt.ylabel('EV sales (World)')
plt.title('Solar power produced (World)')
plt.show()
```

Note: *y-axis* is in millions

```python
Ev_sales_countries = EV_sales.groupby("region")["value"].sum().reset_index(name = "total_sales").sort_values(by = "total_sales")
Ev_sales_countries = Ev_sales_countries[Ev_sales_countries["region"] != "World"]
```

```python
# Scaling of the feature total_sales
from sklearn.preprocessing import StandardScaler 

standerd_scaler = StandardScaler() 
Ev_sales_countries_scaled = standerd_scaler.fit_transform(Ev_sales_countries["total_sales"].values.reshape(-1,1)).flatten()
Ev_sales_countries_scaled
Ev_sales_countries["scaled_values"] = Ev_sales_countries_scaled
```

```python
plt.figure(figsize=(6,9)) 

sns.set_style("whitegrid")
sns.barplot(x = 'region',
            y = 'scaled_values',
            data = Ev_sales_countries[Ev_sales_countries["total_sales"] > Ev_sales_countries["total_sales"].quantile(0.91)], 
            width=0.3,
            palette = "YlGnBu",
            )


plt.xlabel('Entity')
plt.ylabel('EV_sales (scaled)')
plt.title('Solar power produced by entity')
plt.show()
```
